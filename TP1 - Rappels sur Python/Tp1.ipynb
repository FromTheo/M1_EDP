```python
matriceA = np.array([[2,-1,0,0],
                  [-1,2,-1,0],
                  [0,-1,2,-1],
                  [0,0,-1,2]])
matriceB = np.array([[0,1,2,3],
                   [-1,0,1,2],
                   [-2,-1,0,1],
                   [-3,-2,-1,0]])
determinantA = np.linalg.det(matriceA)
determinantB = np.linalg.det(matriceB)
spectreA = np.linalg.eigvals(matriceA)
spectreB = np.linalg.eigvals(matriceB)
print("Le déterminant de A est :", determinantA, "\n Son spectre est : ", spectreA)
print("Le déterminant de B est :", determinantB, "\n Son spectre est : ", spectreB)
```

```python
def det_par_blocs(A,B): #on utilise la formule det(A-(B^t)(A^-1)B) : ok car A est inversible
    B_transpose = np.transpose(B)
    A_inverse = np.linalg.inv(A)
    D = np.dot(B_transpose,np.dot(A_inverse,B))
    C = A - D
    determinant = np.linalg.det(C)
    return determinant 

determinant = det_par_blocs(matriceA,matriceB)
print("Le déterminant de la matrice par blocs C est : ", determinant)
```

## Exercice 3<a id="part3"></a>

```python
def matriceA(n) :
    h = 1/(n+1)   
    diag = 2*np.eye(n)
    diag_inf = -1*np.eye(n, k=-1)
    diag_sup = -1*np.eye(n, k=+1)
    matrice = diag+diag_inf+diag_sup
    return matrice * 1/(h**2)
```

```python
n = 5
h = 1/(n+1) 
A = matriceA(n)
vp = np.linalg.eigvals(A)
print("Matrice A pour n={} : \n {}".format(n,A))

print("Les valeurs propres observées sont : ")
for value in vp :
    print(value)

print("Les valeurs propres théoriques sont : ")
for k in range(1,n+1) : 
    print(2/h**2*(1-np.cos(k*np.pi*h)))
```

```python
def conditionnement(n):
    A = matriceA(n)
    return np.linalg.cond(A,2) #pour la norme 2 

valeurs_n = np.arange(2,100)
cond = [conditionnement(n) for n in valeurs_n]
plt.figure(figsize=(10,6))
plt.loglog(valeurs_n, cond, marker='o', linestyle='-', color='b')
plt.xlabel('Taille de la matrice (n)')
plt.ylabel('Valeur du conditionnement')
plt.title('Conditionnement en norme 2 de la matrice A')
plt.show() 

b = np.ones(n)
sol = np.linalg.solve(A,b)
print("La solution du système linéaire est : ", sol)
```

## Exercice 5<a id="part5"></a>

```python
def f(x) : 
    return np.sin(2*np.pi*x)

def df(x) : 
    return 2*np.pi*np.cos(2*np.pi*x)
```

```python
def d_forward(x,h):
    return (f(x+h)-f(x))/h

def d_backward(x,h) :
    return (f(x)-f(x-h))/h

def d_central(x,h) :
    return (f(x+h)-f(x-h))/(2*h)
```

>**Question 1.**

```python
x_values = np.linspace(-1,1,200)
plt.figure(figsize=(10,6))
h = 0.1 
plt.plot(x_values,df(x_values), label=fr"$f'$") 
plt.plot(x_values,d_forward(x_values,h), label=fr"$d^+$") 
plt.plot(x_values,d_backward(x_values,h), label=fr"$d^-$") 
plt.plot(x_values,d_central(x_values, h), label=fr"$d^0$") 
plt.xlabel('x')
plt.ylabel('values')

plt.legend() 
plt.show()
```

>**Question 2.**

```python
h_values = np.linspace(0.001, 0.1, 200)
max_erreurs = []
for h in h_values :
    erreurs = np.abs(df(x_values)-d_forward(x_values,h))
    max_erreurs.append(np.max(erreurs))

plt.figure(figsize=(10,6))
plt.plot(h_values,max_erreurs)
plt.title('Question 2')
plt.xlabel('h')
plt.ylabel('Erreur max.')
plt.legend() 
plt.show()
```
